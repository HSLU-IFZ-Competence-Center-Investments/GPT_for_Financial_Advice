```python
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin
from pywebcopy import save_webpage
from pywebcopy import save_website
```

```python
# read websites.csv
websites = pd.read_csv('websites.csv', sep='\t')
websites.head()
```

```python
# Trial beautiful soup
def download_page(url, name, count=1, create_folder=True, base_path=None, folder=None):
    response = requests.get(url)
    if response.status_code == 200:
        if create_folder and count == 1:
            folder = os.path.join(base_path, name)
            if not os.path.exists(folder):
                os.makedirs(folder)

        filename = os.path.join(folder, str(count))  # Filename doesn't need URL basename
        # print(f"Downloading {url} to {filename}")
        #filename = filename.replace("\\", "")

        try:
            with open(filename, 'w') as f:
                f.write(response.text)

            soup = BeautifulSoup(response.content, 'html.parser')
            links = soup.find_all('a', href=True)
            # keep only 25 links
            # print(len(links))
            links = links[:25]

            for link in links:
                href = link['href']
                if count == 25:
                    break
                if href.startswith('http'):
                    count += 1
                    download_page(href, name, count, create_folder=False, base_path=base_path, folder=folder)

        except (PermissionError, IOError) as e:
            print(f"Error saving file: {e}")

        except UnicodeEncodeError as e:
            print(f"Error encoding file: {e}")

        # except InvalidChunkLength as e:
        #     print(f"Error encoding file: {e}")

    else:
        print(f"Failed to download {url}")
```
